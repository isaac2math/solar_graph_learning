```r
library(readr)
library(bnlearn)
library(ggplot2)
library(dplyr)
```

```r
rm(list = ls())
linTB <- read_csv("House2010_linear.csv")
logTB <- read_csv("House2010_log.csv")
```

---

# score-based learning on log

# Structure of Baths

```r
data3    = select(logTB, logRent, Baths, logPrice)
net3     = empty.graph(names(data3))
arc.set3 = matrix(c("Baths"   , "logPrice",
                    "Baths"   , "logRent", 
                    "logPrice", "logRent"), 
                  ncol = 2, byrow = TRUE, 
                  dimnames = list(NULL, c("from", "to")))
arcs(net3) = arc.set3
plot(net3)
```

```r
print(paste("BGE score : ", score(net3, data3, type = "bge"  )))
print(paste("BIC score : ", score(net3, data3, type = "bic-g")))
print(paste("AIC score : ", score(net3, data3, type = "aic-g")))
```

```r
net4     = empty.graph(names(data3))
arc.set4 = matrix(c("Baths"   , "logPrice",
                    "logPrice", "logRent"), 
                  ncol = 2, byrow = TRUE, 
                  dimnames = list(NULL, c("from", "to")))
arcs(net4) = arc.set4
plot(net4)
```
